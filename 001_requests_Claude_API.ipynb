## Step 4: Sending Your First Message to Claude

This is the moment everything has been building toward — we're actually sending a message to Claude and getting a response back. Let's break down exactly what's happening.

### What is a function call?

When you write `client.messages.create(...)`, you're calling a **function** — a named action that does something when you trigger it. Think of a function like a vending machine: you put in specific inputs, press a button, and get something back. The inputs you provide are called **parameters** (or **arguments**).

The `client.messages.create()` function is provided by the Anthropic library. It sends your message to Claude over the internet and returns Claude's response.

### What are the parameters?

Everything inside the parentheses `(...)` is a parameter — a piece of information the function needs to do its job:

- **`model=model`** — Tells Claude which version to use. This refers to the `model` variable we set in the previous cell (`"claude-sonnet-4-6"`).
- **`max_tokens=1000`** — Sets a limit on how long Claude's response can be. A **token** is roughly ¾ of a word, so 1000 tokens is about 750 words. This prevents runaway responses and controls cost.
- **`messages=[...]`** — This is the conversation history you're sending to Claude. It's a **list** (notice the square brackets `[...]`) of one or more messages.

### What is a list?

A **list** is an ordered collection of items, written with square brackets `[...]`. Each item is separated by a comma. In this case, our list contains one item — a single message.

### What is the message item?

Each message in the list is a **dictionary** — a set of key-value pairs wrapped in curly braces `{...}`. Think of it like a form with labeled fields:

- **`"role"`** — Who is speaking. `"user"` means it's a message from the human. Later you'll also see `"assistant"` for Claude's replies.
- **`"content"`** — The actual text of the message. This is the question or instruction you're sending.

### What does `message =` do?

The `message =` at the start stores Claude's entire response in a variable called `message`. Claude sends back a lot more than just the text — it includes metadata like which model was used and how many tokens were consumed. We'll access just the text in the next cell.

````python
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role" : "user",
            "content" : "What is quantum computing? Answer in one sentence"
        }
    ]
)
````

## Step 5: Reading Claude's Response

After the previous cell sent your message and stored Claude's full response in the `message` variable, this cell reaches in and pulls out just the text you care about.

### What is `message.content`?

Claude's response is structured like a package with several compartments. The `message` variable holds the whole package. Using a dot `.` is how we access what's inside it — think of it like opening a folder. `message.content` reaches into the response and gets the **content** section, which is a list of response blocks.

### Why is it a list?

Claude can technically return multiple blocks of content in one response (for example, a mix of text and other data types). Even when there's just one response, it still comes back as a list. That's why we need to specify *which* item in the list we want.

### What does `[0]` mean?

Square brackets after a list let you pick a specific item from it by its **index** — its position number. In Python, counting always starts at `0`, not `1`. So `[0]` means "give me the first item." Since Claude almost always returns one block of text, we want item `0`.

### What does `.text` do?

Once we've grabbed the first content block with `[0]`, we use `.text` to access just the plain text string inside it — stripping away all the surrounding structure.

### Putting it all together

`message.content[0].text` reads as: *"From the response, get the content list, take the first item, and give me its text."*

## Step 6: Building a Multi-Turn Conversation

So far we've sent a single message and got a single reply. But real conversations go back and forth — you ask something, Claude responds, you follow up, and so on. This cell sets up the tools we need to make that work.

### What is a multi-turn conversation?

Claude doesn't have memory between separate requests. Every time you call `client.messages.create()`, it's a brand new interaction unless you manually pass in the conversation history. A **multi-turn conversation** means we keep a running list of all the messages exchanged so far — both yours and Claude's — and send the entire history with each new request. That way Claude always has the context of what's been said.

### What is a function definition?

Up until now we've been *calling* functions that already exist (like `client.messages.create()`). Here, we're *defining* our own. The keyword **`def`** tells Python "I'm about to describe a reusable action." The structure looks like this:

```python
def function_name(parameter1, parameter2):
    # code that runs when you call this function
```

Once defined, you can call these functions anywhere below in the notebook, just like any built-in function.

### The three helper functions

**`add_user_message(messages, text)`**

This function takes two parameters — a list of messages and the text you want to add — and appends a new user message to the list. It creates a dictionary with `"role": "user"` and your text as the content, then uses `.append()` to add it to the end of the messages list.

**`add_assistant_message(messages, text)`**

This does the same thing but for Claude's replies. When Claude responds, we save its response back into the conversation history so it remembers what it said. The `"role"` here is `"assistant"`.

**`chat(messages)`**

This is the workhorse function. It takes the full conversation history, sends it to Claude using `client.messages.create()`, and **returns** just the text of Claude's reply. The keyword `return` means "send this value back to whoever called me" — so when you call `chat(messages)`, you get Claude's response text back as a result.

> 💡 Notice that `chat()` uses `return` while the other two functions don't. That's because `add_user_message` and `add_assistant_message` modify the list directly (they don't need to hand anything back), whereas `chat()` needs to give you Claude's response to work with.

````python
# Multi-Turn Conversation: Helper functions

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text
````

## Step 7: Having a Multi-Turn Conversation

Now we put the helper functions from the previous cell to work. This cell demonstrates a real back-and-forth exchange with Claude — two questions, with full context carried through.

### Starting with an empty list

```python
messages = []
```

We begin with a completely empty list. This is our conversation history — it starts blank and grows as the conversation unfolds. The square brackets `[]` with nothing inside them create an empty list.

### Turn 1: Asking the first question

```python
add_user_message(messages, "Define quantum computing in one sentence")
answer = chat(messages)
add_assistant_message(messages, answer)
```

These three lines represent one complete turn:

1. We add our question to the messages list as a user message
2. We send the full list to Claude and store its response in a variable called `answer`
3. We add Claude's response back into the messages list as an assistant message

After these three lines, `messages` contains two items — our question and Claude's answer.

### Turn 2: Asking a follow-up

```python
add_user_message(messages, "Write another sentence")
final_answer = chat(messages)
```

Now we add a follow-up. Notice how vague it is — "Write another sentence" doesn't specify what topic. Claude can make sense of this *only because* the full conversation history is being sent along with it. When `chat(messages)` runs this time, it sends all three messages: our first question, Claude's first answer, and now this follow-up. Claude uses that context to understand what we're asking.

### What is in `final_answer`?

`final_answer` contains Claude's response to the follow-up question — as plain text, ready to read. Because the full conversation history (our first question, Claude's first answer, and the follow-up) was sent along with the second request, Claude had all the context it needed to give a coherent reply.

> 💡 The key insight of this cell: Claude doesn't remember anything on its own. *You* are responsible for keeping track of the conversation and passing the full history every time. The `messages` list is that memory.

````python
# Multi-Turn Conversation:

# Start with an empty message list
messages = []

# Add the initial user question
add_user_message(messages, "Define quantum computing in one sentence")

# Get Claude's response
answer = chat(messages)

# Add Claude's response to the conversation history
add_assistant_message(messages, answer)

#Add a follow-up question
add_user_message(messages, "Write another sentence")

# Get the follow-up response with full context
final_answer = chat(messages)
final_answer
````

## Step 10: Building a Chatbot — Adding a Clean Exit

This is the final and most complete version of the chatbot. It's almost identical to the previous cell, with one important addition: a way to exit the loop gracefully by typing `q` or `quit` instead of having to force-stop it.

### What's new?

```python
if user_input.lower() in {"q", "quit"}:
    break
```

These two lines are the only addition. Let's break them down:

**`if`** — This is a **conditional statement**. It checks whether something is true, and only runs the indented code beneath it if the condition is met. Think of it as "if this is true, do this."

**`user_input.lower()`** — `.lower()` converts whatever the user typed to all lowercase before checking it. This means `"Quit"`, `"QUIT"`, and `"quit"` all work the same way — the check isn't case-sensitive.

**`in {"q", "quit"}`** — The curly braces `{...}` here create a **set** — a collection of values to check against. `in` checks whether the user's input matches any value in that set. So typing either `q` or `quit` will trigger the exit.

**`break`** — This is a special keyword that immediately exits the loop, no matter what. Instead of force-stopping the program from outside, `break` lets the program stop itself cleanly from within.

### Why does this matter?

In the previous cell, the only way to stop the chatbot was to interrupt execution from the menu — an abrupt, external stop. With `break`, the program stops on its own terms, which is cleaner and more like how real applications work.

> 💡 This is the version of the chatbot you'll want to use when experimenting. Just type `q` or `quit` when you're done with the conversation and the loop will exit on its own.

````python
# Make an initial list of messages
messages = []

# Use a ‘while True’ loop to run the chatbot forever
while True:
    # Get user input
    user_input = input("> ")

    if user_input.lower() in {"q", "quit"}:
        break

    # Add user input to the list of messages
    add_user_message(messages, user_input)

    # Call Claude with the 'chat' function
    answer = chat(messages)

    # Add generated text to the list of messages
    add_assistant_message(messages, answer)

    # Print the generated text
    print(answer)

````
